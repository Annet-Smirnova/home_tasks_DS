1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,50)
f= 10*x - 5
#Похідну фінкції буде константа 10 на всьому проміжку
df = np.full_like(x, 10)
# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10*x - 5', color='blue')
plt.plot(x, df, label="f'(x) = 10", color='green')
plt.title("Функція та її похідна")
plt.xlabel("Години навчання (x)")
plt.ylabel("Оцінка")
plt.show()
```

За графіком розуміємо що кожна година навчання додає до оцінки 10 балів

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
x = sp.symbols('x')
xs = sp.symbols('xs')
```

```python
#f(x)=sin(x)
f_sin=sp.sin(x)
derivative_f_sin = sp.diff(f_sin, x)


#f(xs)=ln(xs)
f_ln=sp.log(xs)
derivative_f_ln = sp.diff(f_ln, xs)


print("Похідна f(x)=sin(x):", derivative_f_sin)
print("Похідна f(xs)=ln(xs):", derivative_f_ln)
```

```python
f_sin_func = sp.lambdify(x, f_sin, modules=['numpy'])
df_sin_func = sp.lambdify(x, derivative_f_sin, modules=['numpy'])
x = np.linspace(0,10,200)
y = f_sin_func(x)
dy = df_sin_func(x)
# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='f(x) = sin(x)', color='blue')
plt.plot(x, dy, label="f'(x) = cos(x)", color='green', linestyle='--')
plt.title("Графік функції sin(x) та її похідної")
plt.xlabel("x")
plt.ylabel("Значення")
plt.legend()
plt.grid(True)
plt.show()
```

```python
f_ln_func = sp.lambdify(xs, f_ln, modules=['numpy'])
df_ln_func = sp.lambdify(xs, derivative_f_ln, modules=['numpy'])
xs = np.linspace(0.01,10,200)
y = f_ln_func(xs)
dy = df_ln_func(xs)
# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(xs, y, label='f(xs) = ln(xs)', color='blue')
plt.plot(xs, dy, label="f'(xs) = 1/xs", color='green', linestyle='--')
plt.title("Графік функції ln(xs) та її похідної")
plt.xlabel("xs")
plt.ylabel("Значення")
plt.legend()
plt.grid(True)
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
x = sp.symbols('x')
```

```python
f_x=x**2 - 5 * x + 1
derivative_f_x = sp.diff(f_x, x)
print("Похідна f(x):", derivative_f_x)
```

```python
f_x = sp.lambdify(x, f_x, modules=['numpy'])
df= sp.lambdify(x, derivative_f_x, modules=['numpy'])
x = np.linspace(0,100,100)
y = f_x(x)
dy = df(x)
# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='f(x)=x^2−5x+1', color='blue')
plt.plot(x, dy, label="f(x): 2*x - 5", color='green', linestyle='--')
plt.title("Графік функції   f(x)=x^2−5x+1 та її похідної")
plt.xlabel("кількість вироблених одиниць")
plt.ylabel("Значення")
plt.legend()
plt.grid(True)
plt.show()
```
